```python
import seaborn as sns
import matplotlib.pyplot as plt

# LOAD THE "tips" DATA SET .

plt.style.use("fivethirtyeight")
df = sns.load_dataset("tips")
df
```

```python
# PLOT THE STRIP PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING catplot() .

sns.catplot ( x="day" , y="tip" , kind="strip" , data=df )
```

```python
# PLOT THE STRIP PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING stripplot() .

sns.stripplot ( x="day" , y="tip" , data=df )
```

```python
# PLOT THE SWARM PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING catplot() .

sns.catplot ( x="day" , y="tip" , kind="swarm" , data=df )
```

```python
''' 
PLOT THE SWARM PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "tip" AS Y-ASIS 
USING hue PARAMETER AS COLUMN "sex" USING catplot() . 
'''

sns.catplot ( x="day" , y="tip" , kind="swarm" , hue="sex" , data=df )
```

```python
''' 
PLOT THE SWARM PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "tip" AS Y-ASIS 
USING hue PARAMETER AS COLUMN "day" USING catplot() .
'''

sns.catplot ( x="day" , y="tip" , kind="swarm" , hue="day" , data=df )
```

```python
# PLOT THE SWARM PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING swarmplot() .

sns.swarmplot ( x="day" , y="tip" , data=df )
```
