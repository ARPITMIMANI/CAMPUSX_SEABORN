```python
import seaborn as sns
import matplotlib.pyplot as plt

# LOAD THE "tips" DATA SET .

plt.style.use("fivethirtyeight")
df = sns.load_dataset("tips")
df
```

```python
# PLOT THE JOINT PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING jointplot() .

sns.jointplot ( x="total_bill" , y="tip" , data=df )
```

```python
# PLOT THE JOINT PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING jointplot() WITH kind AS "scatter" .

sns.jointplot ( x="total_bill" , y="tip" , kind="scatter" , data=df )
```

```python
# PLOT THE JOINT PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING jointplot() WITH kind AS "hex" .

sns.jointplot ( x="total_bill" , y="tip" , kind="hex" , data=df )
```

```python
# PLOT THE JOINT PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING jointplot() WITH kind AS "kde" .

sns.jointplot ( x="total_bill" , y="tip" , kind="kde" , data=df )
```

```python
# PLOT THE JOINT PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING jointplot() WITH kind AS "reg" .

sns.jointplot ( x="total_bill" , y="tip" , kind="reg" , data=df )
```
