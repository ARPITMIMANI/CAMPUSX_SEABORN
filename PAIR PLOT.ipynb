```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# LOAD THE "IRIS_DATA.csv" DATA SET .

plt.style.use("fivethirtyeight")
df = pd.read_csv("IRIS_DATA.csv")
df
```

```python
# PLOT THE PAIR PLOT ON DATA SET df USING pairplot() .

sns.pairplot ( data=df )
```

```python
# PLOT THE PAIR PLOT ON DATA SET df USING hue PARAMETER AS COLUMN "species" USING pairplot() .

sns.pairplot ( data=df , hue="species" )
```
