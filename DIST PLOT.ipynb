```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# LOAD THE "TITANIC_DATA.csv" DATA SET .

plt.style.use("fivethirtyeight")
df =pd.read_csv("TITANIC_DATA.csv")
df
```

```python
# CLEAN THE DATA SET WITH fillna() BY FILLING NULL VALUES OF COLUMN "age" AS mean() VALUES .

df["age"].fillna ( df["age"].mean() , inplace=True )
df
```

```python
# PLOT THE DIST PLOT FOR COLUMN "age" USING distplot() .

sns.distplot ( df["age"] )
```

```python
# PLOT THE DIST PLOT FOR COLUMN "age" ONLY THE hist USING distplot() .

sns.distplot ( df["age"] , kde=False )
```

```python
# PLOT THE DIST PLOT FOR COLUMN "age" ONLY THE hist USING distplot() .

sns.distplot ( df["age"] , kde=False , bins=10 )
```

```python
# PLOT THE DIST PLOT FOR COLUMN "age" ONLY THE hist USING distplot() .

sns.distplot ( df["age"] , kde=False , bins=80 )
```

```python
# PLOT THE DIST PLOT FOR COLUMN "age" ONLY THE kde USING distplot() .

sns.distplot ( df["age"] , hist=False )
```

```python
# PLOT THE DIST PLOT FOR COLUMN "age" ONLY THE rug USING distplot() .

sns.distplot ( df["age"] , hist=False , kde=False , rug=True )
```

```python
# PLOT THE DIST PLOT OF THE COLUMN "age" WITH COLUMN "survived" = 1.0 ONLY USING distplot() .

sns.distplot ( df [ df["survived"] == 1 ] ["age"] )
```

```python
# COMPARE THE DIST PLOT OF THE COLUMN "age" WITH COLUMN "survived" = 1.0 AND COLUMN "survived" = 0.0 ONLY USING distplot() .

sns.distplot ( df [ df["survived"] == 1 ] ["age"] )
sns.distplot ( df [ df["survived"] == 0 ] ["age"] )
```
