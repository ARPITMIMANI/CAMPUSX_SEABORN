```python
import seaborn as sns
import matplotlib.pyplot as plt

# LOAD THE "tips" DATA SET .

plt.style.use("fivethirtyeight")
df = sns.load_dataset("tips")
df
```

```python
# PLOT THE BAR PLOT FOR COLUMN "smoker" AS X-AXIS AND COLUMN "tip" AS Y-AXIS USING catplot() .

sns.catplot ( x="smoker" , y="tip" , kind="bar" , data=df )
```

```python
# PLOT THE BAR PLOT FOR COLUMN "smoker" AS X-AXIS AND COLUMN "total_bill" AS Y-AXIS USING catplot() .

sns.catplot ( x="smoker" , y="total_bill" , kind="bar" , data=df )
```

```python
''' 
PLOT THE BAR PLOT FOR COLUMN "smoker" AS X-AXIS AND COLUMN "total_bill" AS Y-AXIS USING catplot() 
USING hue PARAMETER AS COLUMN "sex".
'''

sns.catplot ( x="smoker" , y="total_bill" , hue="sex" , kind="bar" , data=df )
```

```python
import numpy as np
sns.catplot ( x="smoker" , y="total_bill" , hue="sex" , kind="bar" , estimator=np.mean , data=df )
```

```python
import numpy as np
sns.catplot ( x="smoker" , y="total_bill" , hue="sex" , kind="bar" , estimator=np.median , data=df )
```

```python
import numpy as np
sns.catplot ( x="smoker" , y="total_bill" , hue="sex" , kind="bar" , estimator=np.std , data=df )
```

```python
import numpy as np
sns.catplot ( x="smoker" , y="total_bill" , hue="sex" , kind="bar" , estimator=np.var , data=df )
```
