```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# LOAD THE "FLIGHTS_DATA.csv" DATA SET .

plt.style.use("fivethirtyeight")
df = pd.read_csv("FLIGHTS_DATA.csv")
df
```

```python
'''
CREATE A PIVOT TABLE WITH index AS COLUMN "year" , columns AS COLUMN "month" , values AS COLUMN "passengers" 
AND aggfunc AS sum .
'''

pt = df.pivot_table ( index="year" , columns="month" , values="passengers" , aggfunc="sum" )
pt
```

```python
# PLOT THE HEAT MAP OF PIVOT TABLE CREATED ABOVE .

sns.heatmap ( data=pt )
```

```python
# PLOT THE HEAT MAP OF PIVOT TABLE WITH OUT COLOUR BAR .

sns.heatmap ( data=pt , cbar=False )
```

```python
# PLOT THE HEAT MAP OF PIVOT TABLE WITH SOME LINE WIDTH .

sns.heatmap ( data=pt , cbar=True , linewidths=0.64 )
```

```python
sns.heatmap ( data=pt , cbar=True , linewidths=1.08 )
```

```python
sns.heatmap ( data=pt , cbar=True , linewidths=2.56 )
```

```python
# PLOT THE HEAT MAP OF PIVOT TABLE WITH VALUES .

sns.heatmap ( data=pt , cbar=True , linewidths=1.28 , annot=True , fmt="d" )
```

```python
# PLOT THE HEAT MAP OF PIVOT TABLE WITH DIFFERENT COLOUR MAPS  .

sns.heatmap ( data=pt , cbar=True , linewidths=1.28 , annot=True , fmt="d" , cmap="plasma" )
```
