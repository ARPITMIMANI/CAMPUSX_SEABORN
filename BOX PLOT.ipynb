```python
import seaborn as sns
import matplotlib.pyplot as plt

# LOAD THE "tips" DATA SET .

plt.style.use("fivethirtyeight")
df = sns.load_dataset("tips")
df
```

```python
# PLOT THE BOX PLOT FOR COLUMN "tip" USING boxplot() .

sns.boxplot ( df["tip"] )
```

```python
# PLOT THE BOX PLOT FOR COLUMN "total_bill" USING boxplot() .

sns.boxplot ( df["total_bill"] )
```

```python
# PLOT THE BOX PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "total_bill" AS Y-AXIS USING catplot() .

sns.catplot ( x="day" , y="total_bill" , kind="box" , data=df )
```

```python
''' 
PLOT THE BOX PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "total_bill" AS Y-AXIS USING catplot() 
USING hue PARAMETER AS COLUMN "sex".
'''

sns.catplot ( x="day" , y="total_bill" , hue="sex" , kind="box" , data=df )
```
