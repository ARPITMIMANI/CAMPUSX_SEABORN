```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# LOAD THE "FLIGHTS_DATA.csv" DATA SET .

plt.style.use("fivethirtyeight")
df = pd.read_csv("FLIGHTS_DATA.csv")
df
```

```python
'''
CREATE A PIVOT TABLE WITH index AS COLUMN "year" , columns AS COLUMN "month" , values AS COLUMN "passengers" 
AND aggfunc AS sum .
'''

pt = df.pivot_table ( index="year" , columns="month" , values="passengers" , aggfunc="sum" )
pt
```

```python
# PLOT THE CLUSTER MAP OF PIVOT TABLE CREATED ABOVE .

sns.clustermap ( data=pt )
```

```python
# PLOT THE CLUSTER MAP OF PIVOT TABLE WITH DIFFERENT metric  .

sns.clustermap ( data=pt , metric="correlation" )
```

```python
# PLOT THE CLUSTER MAP OF PIVOT TABLE WITH NORMAIZED DATA  .

sns.clustermap ( data=pt , metric="correlation" , z_score=0 )
```

```python
# PLOT THE CLUSTER MAP OF PIVOT TABLE WITH VALUES .

sns.clustermap ( data=pt , metric="correlation" , z_score=0 , annot=True )
```

```python
# PLOT THE CLUSTER MAP OF PIVOT TABLE WITH COLUMN CLUSTER .

sns.clustermap ( data=pt , metric="correlation" , z_score=0 , annot=True , row_cluster=False )
```

```python
# PLOT THE CLUSTER MAP OF PIVOT TABLE WITH ROW CLUSTER .

sns.clustermap ( data=pt , metric="correlation" , z_score=0 , annot=True , col_cluster=False )
```
