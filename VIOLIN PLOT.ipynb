```python
import seaborn as sns
import matplotlib.pyplot as plt

# LOAD THE "tips" DATA SET .

plt.style.use("fivethirtyeight")
df = sns.load_dataset("tips")
df
```

```python
# PLOT THE VIOLIN PLOT FOR COLUMN "total_bill" USING violinplot() .

sns.violinplot ( df["total_bill"] )
```

```python
# PLOT THE VIOLIN PLOT FOR COLUMN "tip" USING violinplot() .

sns.violinplot ( df["tip"] )
```

```python
# PLOT THE VIOLIN PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "total_bill" AS Y-AXIS USING catplot() .

sns.catplot ( x="day" , y="total_bill" , kind="violin" , data=df )
```

```python
''' 
PLOT THE VIOLIN PLOT FOR COLUMN "day" AS X-AXIS AND COLUMN "total_bill" AS Y-AXIS USING catplot() 
USING hue PARAMETER AS COLUMN "sex".
'''

sns.catplot ( x="day" , y="total_bill" , hue="sex" , kind="violin" , data=df )
```

```python
sns.catplot ( x="day" , y="total_bill" , hue="sex" , kind="violin" , split=True , data=df )
```
