```python
import seaborn as sns
import matplotlib.pyplot as plt

# LOAD THE "tips" DATA SET .

plt.style.use("fivethirtyeight")
df = sns.load_dataset("tips")
df
```

```python
# PLOT THE SCATTER PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING relplot() .

sns.relplot ( x="total_bill" , y="tip" , data=df )
```

```python
sns.relplot ( x="total_bill" , y="tip" , data=df , kind="scatter" )
```

```python
# PLOT THE LINE PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS USING relplot() .

sns.relplot ( x="total_bill" , y="tip" , data=df , kind="line" )
```

```python
''' 
PLOT THE SCATTER PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS 
USING hue PARAMETER AS COLUMN "smoker" USING relplot() .
'''

sns.relplot ( x="total_bill" , y="tip" , data=df , hue="smoker" )
```

```python
'''
PLOT THE SCATTER PLOT BETWEEN COLUMN "total_bill" AND COLUMN "tip" USING hue PARAMETER AS COLUMN "smoker" 
AND USING style PARAMETER AS COLUMN "sex" USING relplot() .
'''

sns.relplot ( x="total_bill" , y="tip" , data=df , hue="smoker" , style="sex" )
```

```python
'''
PLOT THE SCATTER PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS 
USING hue PARAMETER AS COLUMN "smoker" , USING style PARAMETER AS COLUMN "sex" AND 
USING size PARAMETER AS COLUMN "size" USING relplot() .
'''

sns.relplot ( x="total_bill" , y="tip" , data=df , hue="smoker" , style="sex" , size="size" )
```

```python
sns.relplot ( x="total_bill" , y="tip" , data=df , hue="smoker" , style="sex" , size="size" , sizes=(60,120) )
```

```python
'''
PLOT THE SCATTER PLOT FOR COLUMN "total_bill" AS X-AXIS AND COLUMN "tip" AS Y-ASIS AND 
USING hue PARAMETER AS COLUMN "smoker" USING scatterplot() .
'''

sns.scatterplot ( x="total_bill" , y="tip" , data=df , hue="smoker" ) 
```
